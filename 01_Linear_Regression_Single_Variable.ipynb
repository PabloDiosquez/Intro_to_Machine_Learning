```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning/area_price.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Area (sqr ft)')
plt.ylabel('Price (US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
```

```python
new_data = pd.read_csv('/content/drive/MyDrive/Machine Learning/new_area.csv')
new_data.head(4)
```

```python
prices = reg.predict(new_data[['area']])
prices
```

```python
m = reg.coef_
b = reg.intercept_

x = 3200
y = m * x + b
y
```

```python
%matplotlib inline
plt.xlabel('Area (sqr ft)')
plt.ylabel('Price (US$)')
plt.scatter(df.area, df.price, color='red', marker='+')
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```
